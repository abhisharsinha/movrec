```python
import pandas as pd
```

```python
import numpy as np
```

```python
from pymongo import MongoClient
```

```python
client = MongoClient("localhost",27017)
```

```python
ratings = pd.read_csv("Data/ratings.dat", sep="::", header=None, names=["user_id", "movie_id", "rating", "timestamp"])
```

```python
ratings.drop(axis=1, labels=["timestamp"], inplace=True)
```

```python
ratings.head()
```

```python
ratings.dtypes
```

```python
ratings = ratings.applymap(str)
```

```python
db = client.movrec
```

```python
ratings[ratings["movie_id"]=="858"].head()
```

```python
movie_ratings = dict()
```

```python
for i in ratings.index:
    try:
        movie_ratings[ratings.loc[i, "movie_id"]].append({ratings.ix[i, "user_id"]: ratings.ix[i, "rating"]})
    except:
        movie_ratings[ratings.loc[i, "movie_id"]] = [{ratings.ix[i, "user_id"]: ratings.ix[i, "rating"]}]
```

```python
movie_ratings["858"]
```

```python
movieratings = []
for i,j in movie_ratings.items():
    movieratings.append({"movie_id":i, "ratings":j})
```

```python
type(movieratings)
```

```python
movieratings[:10]
```

```python
result = db.MovieRatings.insert_many(movieratings)
```

```python
movies = pd.read_csv("Data/movies.dat", sep=";", index_col=0, skiprows=1,names=["title", "genres", "movie_id", "imdb_id"])
```

```python
movies.head()
```

```python
movies = movies.applymap(str)
```

```python
movies = movies.to_dict(orient="records")
```

```python
movies[:5]
```

```python
for i in range(len(movies)):
    movies[i]["genres"] = movies[i]["genres"].split("|")
```

```python
result = db.Movies.insert_many(movies)
```
