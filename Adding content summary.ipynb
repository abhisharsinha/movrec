```python
def fetchSummary(imdb):
    url = "https://www.imdb.com/title/tt" + "0"*(7 - len(str(imdb["imdb_id"]))) + str(imdb["imdb_id"])
    res = requests.get(url)
    soup = BeautifulSoup(res.text)
    try:
        text = soup.find("div").select(".summary_text")[0]
        text = str(text).split(">")[1].split("<")[0].strip()
        db.MovieSummary.insert({"movie_id":imdb["movie_id"], "summary":text})
    except:
        print(imdb["movie_id"])
        print(url)
```

```python
import threading
```

```python
t = [0]*25
for i in range(len(list_of_imdb)):
    t[i%25] = threading.Thread(target=fetchSummary, args=(list_of_imdb[i], ))
    t[i%25].start()
    if (i+1) % 25 == 0:
        for j in range(25):
            t[j].join()
        print(str(i) + " / " + str(len(list_of_imdb)))
```
